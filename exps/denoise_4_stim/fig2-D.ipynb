```python
data = {
    'dropout': ["33.67% ", "36.86%", "44.01%", "52.09%", "68.93%", "71.99%", "78.02%", "83.68%", "89.42%"],
    'DeepImpute': [0.5901, 0.5841, 0.5641, 0.5118, 0.3831, 0.3427, 0.2692, 0.2033, 0.1590],
    'MAGIC': [0.6550, 0.6521, 0.6462, 0.6390, 0.6163, 0.6107, 0.6003, 0.5841, 0.5687],
    'SCALEX':  [0.6400, 0.6458, 0.6501, 0.6551, 0.6540, 0.6476, 0.6425, 0.6350, 0.6113],
    'ComGRN': [0.7103, 0.7289, 0.7158, 0.7188, 0.7133, 0.7158, 0.6723, 0.6681, 0.5732],
}

df = pd.DataFrame(data)
df = df.set_index('dropout')

print(df)
```

```python
# 
fig, ax = plt.subplots(figsize=(10, 4))

for method, color in zip(df.columns, colors):
    ax.plot(df.index, df[method], marker='o', color=color, label=method)


ax.set_xlabel('Dropout Rate')
# ax.set_title('Comparison of Different Methods')
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.ylabel('Cell Corr')
plt.grid(False)

# plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
