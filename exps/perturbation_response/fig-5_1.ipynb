```python
import matplotlib.pyplot as plt
import numpy as np
import scanpy as sc

np.random.seed(1)
plt.rcParams['text.color'] = 'black'
sc.set_figure_params(dpi=100, dpi_save=300, facecolor='white')
plt.style.use('ggplot')
```

```python
# 示例数据
categories = ['ComGRN(Ours)', 'scPreGAN', 'scDisInFact', 'scShift']
values1 = [0.9141, 0.8337, 0.8968, 0.9232]  # 第一个组的数据
values2 = [0.2316, 0.1805, 0.1764, 0.1865]  # 第二个组的数据
x = np.arange(len(categories))  # x轴坐标

plt.figure(figsize=(6, 4))
# 绘制叠加的条形图
bars1 = plt.bar(x, values1, width=0.5, align='center', label='R\u00B2')
bars2 = plt.bar(x, values2, width=0.5, align='center', label='Adj. R\u00B2', bottom=values1)

# 添加标签和图例
plt.xlabel('Methods')
plt.ylabel('Values')
plt.xticks(x, categories, fontsize=10)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

# 在每个条形的顶部添加数字标签
for bar1, bar2 in zip(bars1, bars2):
    plt.text(bar1.get_x() + bar1.get_width() / 2, bar1.get_height() - 0.08,
             f'{bar1.get_height():.4f}', ha='center', va='bottom')
    plt.text(bar1.get_x() + bar1.get_width() / 2, bar1.get_height() + bar2.get_height() - 0.08,
             f'{bar2.get_height():.4f}', ha='center', va='bottom')

# 显示图形
plt.title("The average R\u00B2 of different cell types")
plt.savefig('./fig-5.pdf')

plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# 数据
data = [[57, 46, 47, 48, 44, 54, 79],
        [41, 29, 25, 43, 17, 37, 37],
        [0, 1, 0, 0, 0, 1, 0],
        [23, 12, 13, 10, 3, 5, 25]]

# 方法名称
methods = ['Dendritic', 'B', 'CD4T', 'FCGR3A+Mono', 'NK', 'CD8T', 'CD14+Mono']

# 角度和数值
# 角度和数值
angles = np.linspace(0, 2 * np.pi, len(methods), endpoint=False).tolist()
angles += angles[:1]

# 创建图形
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111, polar=True)

# 绘制雷达图
labels = ['ComGRN(ours)', 'scPreGAN', 'scDisInFact', 'scShift']

values = data[0] + data[0][:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label=labels[0])
ax.fill(angles, values, alpha=0.25)

values = data[1] + data[1][:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label=labels[1])
ax.fill(angles, values, alpha=0.25)

values = data[2] + data[2][:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label=labels[2])
ax.fill(angles, values, alpha=0.25)

values = data[3] + data[3][:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label=labels[3])
ax.fill(angles, values, alpha=0.25)

# 添加标签
# 设置x轴刻度标签间隔
ax.tick_params(axis='x', which='major', pad=20) 
ax.set_xticks(np.linspace(0, 2*np.pi, len(methods), endpoint=False))
ax.set_xticklabels(methods)

# ax.set_yticklabels([1, 2, 3, 4, 5, 6, 7, 8])
ax.legend(loc='upper right', bbox_to_anchor=(1.5, 1.2))
plt.title("The common DEG of different cell type")
# 显示图形

plt.savefig('./fig-5_commongene.pdf')

plt.show()
```

```python
import pandas as pd

plt.rcParams['text.color'] = 'black'
plt.style.use('ggplot')

# 创建数据
mean = {
    'metrics': ['AUPRC', 'AUROC', 'EP', 'PCC'],
    'scDisInFact': [0.141158013, 0.497849569, 0.126573427, -0.002404519],
    'ComGRN(Ours)': [0.580942942, 0.813526921, 0.55, 0.529585621],
}

std = {
    'metrics': ['AUPRC', 'AUROC', 'EP', 'PCC'],
    'scDisInFact': [0.003562735, 0.009321541, 0.015532244, 0.002426438],
    'ComGRN(Ours)': [0.08294113, 0.034215698, 0.052377735, 0.072568346],
}

# 创建数据框
df = pd.DataFrame(mean)
df = df.set_index('metrics')

vf = pd.DataFrame(std)
vf = vf.set_index('metrics')
```

```python
# 绘制条形图
dropouts = df.index.to_series()

# Create a bar chart
fig, ax = plt.subplots(figsize=(8, 6))

for i, label in enumerate(df.columns):
    # print(label)
    ax.bar([x + (i- 0.5) * 0.4 for x in range(len(dropouts))], df[label], yerr=vf[label], width=0.4, label=label)
    # ax.set_xticklabels(df.index, rotation=45, ha='right')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    
ax.xaxis.label.set_color('black')
ax.yaxis.label.set_color('black')
ax.set_xticks(range(len(dropouts)))
ax.set_xticklabels(dropouts)
# ax.set_title('Comparison of Different Methods')
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.xlabel('Metrics')
plt.ylabel('Value')
plt.grid(False)
# plt.xticks(rotation=45)
plt.tight_layout()
plt.title("The different Metrics of CKG detection")
plt.savefig('./fig-5_CKG.pdf')
plt.show()
```
