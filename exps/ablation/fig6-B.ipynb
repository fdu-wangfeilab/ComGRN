```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import scanpy as sc

plt.rcParams['text.color'] = 'black'
sc.set_figure_params(dpi=100, facecolor='white')
plt.style.use('ggplot')

colors = ['#17BECF', '#348ABD', '#8EBA42', '#E24A33']
```

```python
data = {
    'ComGRN w/o sparse': 0.568588049,
    'ComGRN w/o concept': 0.604233518,
    'ComGRN w/o adv': 0.546826942,
    'ComGRN':0.613237521 ,
}

methods = list(data.keys())
scores = list(data.values())
```

```python
# 设置 Nature 配色方案
# colors = ['#E64B35', '#4DBBD5', '#00A087', '#3C5488', '#F39B7F', '#8491B4', '#91D1C2', '#DC0000', '#7E6148', '#B09C85']

# 创建图形和轴对象
fig, ax = plt.subplots()

# 绘制横向条形图
bars = ax.barh(methods, scores, color=colors)

for bar, score in zip(bars, scores):
    ax.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, f'{score:.2f}', 
            va='center', ha='right', fontsize=10)
    
# 设置图形标题和标签
ax.set_title('hPancreas')
ax.set_xlabel('scIB Score')
ax.set_ylabel('Methods')

plt.grid(False)
# plt.xticks(rotation=45)
# plt.tight_layout()
plt.show()
# 显示图形
plt.show()
```
