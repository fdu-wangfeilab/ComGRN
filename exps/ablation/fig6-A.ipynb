```python
data = {
    'dropout': ["33.67% ", "36.86%", "44.01%", "52.09%", "68.93%", "71.99%", "78.02%", "83.68%", "89.42%"],
    'ComGRN w/o adv': [0.7109, 0.7114, 0.7029, 0.6894, 0.6894, 0.6253, 0.5930, 0.5462, 0.4488],
    'ComGRN w/o concept': [0.7337, 0.7192, 0.7306, 0.7151, 0.7104, 0.6864, 0.6661, 0.6737, 0.5723],
    'ComGRN w/o sparse': [0.0351, 0.0263, 0.0038, 0.0365, 0.0827, 0.1264, 0.1511, 0.3021, 0.4481],
    'ComGRN': [0.7408, 0.7351, 0.7321, 0.7141, 0.7168, 0.6968, 0.6937, 0.6761, 0.5463],
}

df = pd.DataFrame(data)
df = df.set_index('dropout')

print(df)
```

```python
# 
fig, ax = plt.subplots(figsize=(10, 4))

for method, color in zip(df.columns, colors):
    ax.plot(df.index, df[method], marker='o', color=color, label=method)


ax.set_xlabel('Dropout Rate')
# ax.set_title('Comparison of Different Methods')
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.ylabel('Cell Corr')
plt.grid(False)

# plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
