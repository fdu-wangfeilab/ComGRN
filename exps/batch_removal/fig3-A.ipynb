```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import scanpy as sc

plt.rcParams['text.color'] = 'black'
sc.set_figure_params(dpi=100, facecolor='white')
plt.style.use('ggplot')

colors = ['#F39B7F', '#17BECF', '#348ABD', '#988ED5', '#777777', '#FBC15E', '#F4A582', '#80B1D3', '#8EBA42', '#FFB5B8', '#8491B4', '#E24A33']
```

```python
data = {
    "raw": 0.3503,
    "DeepImpute": 0.4985,
    "Bis": 0.5004,
    "MAGIC": 0.5222,
    "SAVERX": 0.4977,
    "scVI": 0.5018,
    "MNNCorrect": 0.5657,
    "Scanorama": 0.5644,
    "scPSM": 0.4944,
    "SCALEX": 0.5637,
    "Beaconet": 0.5700,
    "ComGRN": 0.6132,
}

methods = list(data.keys())
scores = list(data.values())
```

```python
# 设置 Nature 配色方案
# colors = ['#E64B35', '#4DBBD5', '#00A087', '#3C5488', '#F39B7F', '#8491B4', '#91D1C2', '#DC0000', '#7E6148', '#B09C85']

# 创建图形和轴对象
fig, ax = plt.subplots()

# 绘制横向条形图
bars = ax.barh(methods, scores, color=colors)

for bar, score in zip(bars, scores):
    ax.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, f'{score:.2f}', 
            va='center', ha='right', fontsize=10)
    
# 设置图形标题和标签
ax.set_title('hPancreas')
ax.set_xlabel('scIB Score')
ax.set_ylabel('Methods')

plt.grid(False)
# plt.xticks(rotation=45)
# plt.tight_layout()
plt.savefig('bar_chart_1.pdf', dpi=300, bbox_inches='tight')
# 显示图形
plt.show()

```
